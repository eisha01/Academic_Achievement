```python
# task 6 (a)
import statsmodels.api as sm

# Model 1: using hs_gpa as the independent variable
model_1 = sm.OLS(df['CAE_v'], sm.add_constant(df['hs_gpa']),)
result_1 = model_1.fit()

# Model 2: using fy_gpa as the independent variable
model_2 = sm.OLS(df['CAE_v'], sm.add_constant(df['fy_gpa']))
result_2 = model_2.fit()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Residual Plot for Model 1
plt.scatter(result_1.predict(sm.add_constant(df['hs_gpa'])), result_1.resid)
plt.title('Residual Plot for Model 1')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Residual Plot for Model 2
plt.scatter(result_2.predict(sm.add_constant(df['fy_gpa'])), result_2.resid)
plt.title('Residual Plot for Model 2')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Standard Error for Model 1
std_err_1 = np.sqrt(result_1.mse_resid)
print("Standard Error for Model 1:", std_err_1)

# Standard Error for Model 2
std_err_2 = np.sqrt(result_2.mse_resid)
print("Standard Error for Model 2:", std_err_2)

# R-squared for Model 1
r_squared_1 = result_1.rsquared
print("R-squared for Model 1:", r_squared_1)

# R-squared for Model 2
r_squared_2 = result_2.rsquared
print("R-squared for Model 2:", r_squared_2)
```
